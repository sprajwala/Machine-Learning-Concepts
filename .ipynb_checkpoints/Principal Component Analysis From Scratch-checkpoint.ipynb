# Principal Component Analysis From Scratch

```python
import numpy as np
from numpy import linalg as LA
from sklearn.decomposition import PCA
import matplotlib
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

## a) Find sample mean

```python
# Load Data
sample = [[3,2,1],[2,4,5],[1,2,3],[0,2,5]]

sample_mean = np.mean(sample, axis=0)
print("Sample Mean: ",sample_mean)
```

## b) # Zero center the samples

```python
# Center data around mean by subtracting it from all samples
sample_centered = sample - np.mean(sample, axis=0)
sample_centered
```

```python
# visualize that the data is mean centered. 
plt.plot(sample_centered,'o')
```

## # Find data covariance matrix
Covariance indicates the level to which two variables vary together. If we examine N-dimensional samples, $X = [x_1, x_2, ... x_N]^T$, then the covariance matrix element $ C_{ij}$ is the covariance of $x_i$ and $x_j$. The element $C_{ii}$ is the variance of $x_i$.

$$ Cov(X,Y) = \frac{\sum(X_i - \bar{X}) (Y_j - \bar{Y})}{n} $$

```python
# Pass transpose of dataset because numpy's implementation of cov assumes rows to be features
Q = np.cov(sample_centered.T)
print("Sample Covariance: \n",Q)
```

## # Find eigenvalues and eigenvectors of Q

```python
# w is the set of Eigenvalues and v is the set of Eigenvectors
w , v = LA.eig(Q)
print("\n Eigenvalues: \n",w)
print("\n Eigenvectors: \n",v)
```

## c) PCA -  WITHOUT sklearn

```python
# Observe variance along eigenvectors obtained
total = sum(w)
var_exp = [(i / total)*100 for i in sorted(w, reverse = True)]
cum_var_exp = np.cumsum(var_exp)
print(cum_var_exp)
```

We can see that the first two Eigenvectors capture most of the variance in the data. (99.6%)

```python
# Selecting top two eigenvectors
top2_vectors = v[:,0:2]
projection = sample_centered.dot(top2_vectors)
print("PCA Coefficients corresponding to each of the samples: \n",projection)
```

## d) Reconstruction using first two components

```python
sample_reconstructed = projection.dot(top2_vectors.T) + sample_mean
print("\nReconstructed Sample: \n",sample_reconstructed)

# Calculate reconstruction error

reconstruction_error = ((sample_centered - sample_reconstructed)**2)
print("\nReconstruction Error: \n",reconstruction_error)

print("\n Mean Reconstruction Error: \n", mean_squared_error(sample, sample_reconstructed))
```

## c) PCA - using sklearn

```python
pca = PCA(n_components=2)
pca.fit(sample_centered)
sample_projected = pca.fit_transform(sample_centered)

# Plot data against first two principal components
plt.scatter(sample_projected[:, 0], sample_projected[:, 1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Projection on two principal components')
plt.show()
```

```python
# PCA Coefficients

a_1 = pca.components_.T # get a_1 that maximizes variance
print(a_1)
scores_a1 = np.matmul(sample_centered, a_1)
print(scores_a1)
```

```python
print("Sample Projected on first two principal components: \n",sample_projected)
```

## d) Reconstructing from top two principal components

```python
sample_reconstructed = pca.inverse_transform(sample_projected)
```

```python
loss = ((sample_centered - sample_reconstructed) ** 2).mean()
print("Reconstruction Error: ",loss)
```

# K Means Clustering on Senator Dataset

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
```

```python
# Load Data
url_votes = 'https://raw.githubusercontent.com/exemplary-citizen/PCA-and-Senate-Voting-Data/master/senator_pca_problem/senator_data_pca/'
senator_df  =  pd.read_csv(url_votes + 'data_matrix.csv',error_bad_lines=False)
af = pd.read_csv(url_votes + 'politician_labels.txt', header=None)
af["affiliations"] = af[0].str.split().str[-1]

# Remove the unnecessary columns from dataset
senator_df = senator_df.loc[:, ~senator_df.columns.str.contains('^Unnamed')]
senator_df.drop(['bill_type bill_name bill_ID',' missing_votes'],axis=1,inplace=True)

senator_df.head(3)
```

```python
# Flip the dataset to get all senators as rows, ie, each senator corresponding to a data point
df = senator_df.T

# Append the actual affiliations to the dataset
df.insert(0,'name',af[0].values)
df.insert(1,'affiliations',af["affiliations"].values)
df.head(3)
```

```python
X = np.array(df.values[:, 2:], dtype='float64') #transpose to get senators as rows
```

```python
X.shape
```

```python
kmeans_model = KMeans(n_clusters=2, random_state=1).fit(X)
labels = kmeans_model.labels_
centers = kmeans_model.cluster_centers_
centers
```

votes = senator_df <br>
votes["party"] = af[0]

```python
# Check how many senators have been misclassified
print(pd.crosstab(labels, af["affiliations"]))
```

```python
senator_mis_1 = df[(labels == 0) & (df["affiliations"] == "Red")]
senator_mis_2 = df[(labels == 1) & (df["affiliations"] == "Blue")]
print(senator_mis_1["name"])
print(senator_mis_2["name"])
```

## PCA to show what k-means has learnt

```python
# Reduce dimension of dataset to 2 by performing PCA
pca_2 = PCA(2)
plot_columns = pca_2.fit_transform(X)
plot_centers = pca_2.transform(centers)   # Transform cluster centers also using same PCA
plt.scatter(x=plot_columns[:,0], y=plot_columns[:,1], c=af["affiliations"])
plt.scatter(x=plot_centers[:,0], y=plot_centers[:,1], c='g',marker="x",s=200)
plt.show()
```

<ul>
<li> <b> From the above PCA plot we can see clearly spot the two misclassified senators. One red point in the blue cluster and one blue point among the red cluster. <br> <br>
<li> The reason that K Means has misclassified these points might be because these two are much closer to the cluster centers of the opposite party than their own. <br>
We can visually observe this clearly, the two green crosses denote the two cluster centers and we can see that the misclassified points are much closer to the opposite party's cluster center.
</b>
<li> <b> Also K means gives us circular clusters which might be another reason for misclassification

<li> <b> We can qualitatively understand why these points are closer to the opposite cluster by looking at how many times these two senators have abstained their vote showing non-typical party behavior. 
<li> <b> This is shown in the results below for the misclassified senators : Lincoln D. Chafee and E. Benjamin'Ben'Nelson 
    </ul>

```python
print("Voting Pattern for Lincoln D. Chafee(RI)")
print(pd.value_counts(senator_mis_1.iloc[:,2:].values.ravel()))
```

```python
print("Voting Pattern for E. Benjamin'Ben'Nelson")
print(pd.value_counts(senator_mis_2.iloc[:,2:].values.ravel()))
```

<b> We can see that both of them have abstained their vote far more times than they have voted 'for' or 'against'.

# Interpret <i>Places Related Almanac Dataset </i> using PCA

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib
```

## a) Load Data

```python
cols_to_use = ['Climate','HousingCost','HlthCare','Crime','Transp','Educ','Arts','Recreat','Econ']
df = pd.read_csv('places.txt',delim_whitespace=True,header=0,usecols=cols_to_use)
```

```python
df
```

```python
df.shape
```

## b) Scaling Data
Take log10 of all values

```python
def log_10(x):
    return math.log10(x)

df = df.apply(np.vectorize(log_10))
df.head(4)
```

## c) Perform PCA on Data

### Standardize Data
Center around mean

```python
#df = df.apply(lambda x: x-df.mean()) 
df = df - df.mean()
df.head(5)
```

```python
df.shape
```

## d)

### Sigular Value Decomposition

$$ A_{[mxn]} = U_{[mxr]} \sum_{[rxr]} (V_{[nxr]})^T $$

$ A =  $ matrix of shape m x n <br>
where, , m=329 n=9 <br>
$U = $ Left Singular Vectors shape (m x r) r = concepts/topics <br>
$S = $ Singular values along diag sorted in dec order, diag matrix of size (r x r) <br>
$V = $ Right singular vectors, (n x r), r = rank of matrix A

SVD theorem says:
<ol>
    <li> For real matrix this decomposition is always possible </li>
    <li> Always unique </li>
    <li> U and V are column orthonormal (cols have euclidean len =1,ie, sum of squares in each col =1) cols are orthogonal (if you take dot product of different cols you get zero) </li>
    </ol>

```python
u, s, vh = np.linalg.svd(df, full_matrices=False)
eigvals = s**2 / np.sum(s**2)
```

```python
print(u.shape, s.shape, vh.shape)
```

```python
s
```

```python
vh
```

```python
np.argmax(vh, axis=0)
```

```python
# SVD gives us eigenvectors row-wise in the matrix vh
v1 = vh.T[:,0]
v2 = vh.T[:,1]
```

```python
print("\n First Singular Vector v1: \n ",v1)
print("\n Second Singular Vector v2: \n", v2)
```

By looking at maximum absolute value in the eigenvectors, we can determine which feature the particular vector correlates with the most.  <br>
<b> The first vector correlates most to the 7th column Arts. </b> <br>
<b> The second vector correlates 3rd column Healthcare </b>

```python
P = vh.T.dot(df.T)
```

```python
P.T
```

```python
fig = plt.figure(figsize=(8,5))
sing_vals = np.arange(9) + 1
plt.plot(sing_vals, eigvals, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
 

plt.show()
```

```python
from numpy import cov
cov_mat = cov(df.T)
cov_mat
```

## e) Project data onto principal components

## f) Repeating with z score

```python
cols_to_use = ['Climate','HousingCost','HlthCare','Crime','Transp','Educ','Arts','Recreat','Econ']
df = pd.read_csv('places.txt',delim_whitespace=True,header=0,usecols=cols_to_use)
```

```python
# Calculate mean and standard deviation
means = df.mean()
std_dev = df.std(axis=0)
```

$$ z score = \frac{x - \mu}{\sigma} $$

```python
df_z = (df - means) / std_dev
df_z.head(3)
```

### SVD and Principal Components

```python
U, S, VH = np.linalg.svd(df_z, full_matrices=False)
eigvals = s**2 / np.sum(s**2)
```

```python
print(U.shape, S.shape, VH.shape)
```

```python
S
```

```python
VH
```

```python
np.argmax(VH, axis=0)
```

```python
v1 = VH.T[:,0]
v2 = VH.T[:,1]
```

```python
print("\n First Singular Vector v1: \n ",v1)
print("\n Second Singular Vector v2: \n", v2)
```

<b> The first vector correlates most to the 7th column Arts. </b> <br>
<b> The second vector correlates 6th column Education welfare </b>

```python
pca_places_z = PCA(n_components=2)
pca_places_z.fit(df_z)
df_projected_z = pca_places_z.fit_transform(df_z)

# Plot data against first two principal components
plt.scatter(df_projected_z[:,0], df_projected_z[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Projection on two principal components')
```

```python
df_plot = pd.DataFrame({'x':df_projected_z[:, 0], 'y':df_projected_z[:, 1]}, index=df.index.values.tolist() )

fig, ax = plt.subplots(figsize=(15,10))
df_plot.plot('x','y',kind='scatter',ax=ax)

for k, v in df_plot.iterrows():
    ax.annotate(k,v)
    
fig.canvas.draw()
```
